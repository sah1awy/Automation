```python
from pycaret.datasets import get_data
```

```python
get_data('index')
```

```python
dataset = get_data('employee')
```

```python
dataset.info()
```

```python
dataset.nunique()
```

```python
dataset = dataset.sample(5000)
```

```python
data = dataset.sample(frac=0.9)
data_unseen = dataset.drop(data.index)
```

```python
data.reset_index(drop=True,inplace=True)
data_unseen.reset_index(drop=True,inplace=True)
```

```python
print('training Data shape: {}'.format(data.shape[0]))
print('training Data shape: {}'.format(data_unseen.shape[0]))
```

```python
from pycaret.classification import *
```

```python
e = ClassificationExperiment()
```

```python
cat = ['department', 'salary','promotion_last_5years','Work_accident']
```

```python
e.setup(data=data,preprocess=True,train_size=0.8,target='left',categorical_features=cat)
```

```python
e.models()
```

```python
e.compare_models()
```

```python
lg = e.create_model('lightgbm')
```

```python
lg_tunded = e.tune_model(lg)
```

```python
gb = e.create_model('gbc')
```

```python
tuned_gb= e.tune_model(gb)
```

```python
e.plot_model(calibrated_lg,plot='calibration')
```

```python
final_lg = e.finalize_model(lg)
```
